```python
# Render offscreen -- make sure to set the PyOpenGL platform
from utils import render
import glob

meshnames = sorted(glob.glob('data/shapenet_simplified/val/03001627/*/*.ply'))

for idx in range(20):
    color, depth = render.render_mesh(meshnames[idx], dist=1.2, res=(112, 112), color=[1.,1.,1.], intensity=6)
    # Show the images
    import matplotlib.pyplot as plt
    %matplotlib inline
    plt.figure()
    plt.subplot(1,2,1)
    plt.axis('off')
    plt.imshow(color)
    plt.subplot(1,2,2)
    plt.axis('off')
    plt.imshow(depth, cmap=plt.cm.gray_r)
    plt.show()
```
